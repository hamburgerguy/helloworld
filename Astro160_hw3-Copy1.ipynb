```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

f = np.loadtxt('structure_00001.txt', unpack = True)

gradient_ad = f[10]
gradient_rad = f[15]
gradient = f[16]
r = f[1]

plt.figure(figsize = (12,12))
plt.xlabel('r/r_sol', fontsize = 11)
plt.ylabel('log temp grad', fontsize = 11)
plt.title('temp grad as function of radius', fontsize = 15)

plt.plot(r, np.log10(gradient_ad),'red', label = 'Adiabatic Gradient')
plt.plot(r, np.log10(gradient_rad),'green', label = 'Radiation Gradient')
plt.plot(r, np.log10(gradient),'yellow', label = 'Gradient')
plt.legend(loc = 'upper left', fontsize = 10)
```

a) Convection occurs when the radiation gradient is greater than the adiabatic gradient. On this graph this should coincide with when the line representing the gradient is below the adiabatic gradient, which occurs on the left and right of the graph.

```python
tst = np.zeros(len(gradient_ad))

for i in range(1, len(gradient_ad) -1):
    if gradient_ad[i] > gradient_rad[i-1] and gradient_ad[i] < gradient_rad[i+1]:
        tst[i] = 1

```

```python
p = f[4]
P = f[3]
T = f[5]

xprodp = p * tst
xprodP = P * tst
xprodT = T * tst
xprodR = r * tst

def filt(data):
    non_z = []
    
    for i in data:
        if i != 0:
            non_z.append(i)
    return non_z

filterp = filt(xprodp)
filterP = filt(xprodP)
filterT = filt(xprodT)
filterR = filt(xprodR)

print(filterp, 'Resultant densities at the convective layers') 
print(filterP, 'Resultant pressures at the convective layers')
print(filterT, 'Resultant temperatures at the convective layers')
print(filterR, 'distance from center (radius) locations for convective layers')


```
